Posteriormente se pasa el archivo csv a un dataframe con el objetivo de analizar la estructura de la información del archivo que esta separado por ,

```python

df=pd.read_csv('ventas_techventas.csv',
            sep=",", 
            encoding="latin-1")
df.head()
```

```python
#Aqui se toman 2 columnas del dataframe y los convertimos en 2 arreglos de numpy: cantidad y precio y luego se imprimen por pantalla para ver como se guardaron

cantidad=df['cantidad'].values
precio=df['precio_unitario'].values

print(cantidad)

print("precios-------------")
print(precio)
```

```python
#Aqui se detectan cuántos valores faltantes (NaN) hay en las columnas cantidad y precio_unitario.
print(df[['cantidad', 'precio_unitario']].isna().sum())
```

```python
#Aqui se calcula el producto punto entre los dos arreglos cantidad y precio, pero su resultado es nan debido a que los arreglos todavía tienen valores nulos

ingresos_brutos = np.dot(cantidad,precio)

print(ingresos_brutos)
```

Antes de realizar cualquier imputación o cambio en el archivo csv, se crea una copia independiente para poder modificar el archivo sin afectar el DataFrame original

```python
df_ventas=df.copy() #se crea la copia

#Se reemplazan los valores faltantes (NaN) por 0 en las columnas cantidad y precio_unitario porque el 0 
#  significa ausencia de cantidad o de precio.

df_ventas['cantidad']=df_ventas['cantidad'].fillna(0)
df_ventas['precio_unitario']=df_ventas['precio_unitario'].fillna(0)
```

```python
#Una vez imputados los valores nulos, se vuelve a verificar si la base de datos sigue teniendo valores nulos
print(df_ventas[['cantidad', 'precio_unitario']].isna().sum())
```

```python
#Se obtienen las columnas cantidad y precio_unitario del DataFrame  y se convierten en arreglos de numpy para poder realizar operaciones matemáticas

cantidades=df_ventas['cantidad'].values
precios=df_ventas['precio_unitario'].values

#Después se realiza una multiplicación elemento a elemento donde cantidad se multiplica por su precio correspondiente y el resultado es un arreglo donde cada posición representa el ingreso generado por un producto.

ingresos_de_elemento=(cantidades*precios)
print(ingresos_de_elemento)

#Se calcula el producto punto que multiplica los elementos correspondientes y luego suma todos los resultados para obtener el ingreso bruto total.
ingresos_brutos_totales = np.dot(cantidades,precios)

print()
print(f" los ingresos totales son de ${ingresos_brutos_totales}")


#Se suman todos los ingresos calculados mediante la multiplicación elemento a elemento y se compara con el resultado de producto punto, donde si son iguales, arroja true
print("\nSuma de la multiplicación elemento a elemento:")
print(ingresos_de_elemento.sum())

print("\n¿Son iguales?")
print(ingresos_de_elemento.sum() == ingresos_brutos_totales)
```

```python
#Aqui se calculan estadísticas descriptivas de la columna precios para entender cómo se distribuyen los datos.
print(np.mean(precios))
print(np.median(precios))
print(np.std(precios))
print(np.percentile(precios,[25,75]))
```

# Explicación estadistica
Los resultados de las estadísticas descriptivas del bloque anterior muestran cómo se distribuyen los precios del catálogo donde la media (314.08) es bastante alta en comparación con la mayoría de los valores, lo que indica que está influenciada por productos más costosos, sin embargo, la mediana (95.0) representa mejor el precio típico, ya que divide el catálogo en dos partes iguales y refleja que la mayoría de los productos se encuentran alrededor de ese valor. La desviación estándar (469.09) es alta, lo que indica que existe una gran variabilidad entre los precios, es decir, no todos los productos tienen valores similares y hay diferencias importantes dentro del catálogo y finalmente, los percentiles [25, 120] muestran que el 25% de los productos tienen precios iguales o menores a 25, mientras que el 75% están por debajo o igual a 120, lo que confirma que la mayoría de los productos son relativamente económicos y que solo unos pocos tienen precios más altos que elevan el promedio.
